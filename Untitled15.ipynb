#드라이브에 있는 사진경로를 earth로 설정하고 원본을 읽은 후 보여준다.

```python
earth = '/content/drive/MyDrive/earth.jpg'
```

```python
earth = cv2.imread(earth, cv2.IMREAD_COLOR)
```

```python
cv2_imshow(earth)
```

#사진의 크기를 알기 위해 shape을 써서 크기를 파악한다

```python
earth.shape
```

#원본 사진에서 테두리를 검출하기 위해 위에서 알아냈던 크기를 그대로 대입하고 출력한다.

```python
method_canny = cv2.Canny(earth, 740, 492)
cv2_imshow(method_canny)
```
